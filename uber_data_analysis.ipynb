```python
# importing necessary modules and
# reading in the dataset
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
from csv import reader

open_file = open("C:/Users/piter/OneDrive/Desktop/UberData/Rider/trips_data.csv", encoding = 'utf-8')
read_file = reader(open_file)
uber_data = pd.DataFrame(read_file)

uber_data
```

```python
# removing all columns which will not be used for analysis
uber_data.drop([1, 5, 6, 7, 9, 10, 11], axis = 1, inplace = True)
uber_data
```

```python
# remove all rows where trip status is either unfulfilled or driver_cancelled
uber_data.drop(uber_data[(uber_data[2] != "COMPLETED")].index, inplace = True)
uber_data
```

```python
# converting currencies from pln to gbp
# 1 GBP is roughly 5 PLN so to get correct values
# divide all PLN values by 5
uber_data[13] = pd.to_numeric(uber_data[13])
# uber_data.loc[(uber_data[14] != 'GBP'), uber_data[13]] = uber_data[13] / 5
uber_data[13] = np.where(uber_data[14] != 'GBP', uber_data[13] / 5, uber_data[13])
uber_data[14] = np.where(uber_data[14] != 'GBP', 'GBP', uber_data[14])
uber_data
```

```python
# remove +0000 UTC from all dates
uber_data[3] = uber_data[3].map(lambda x: x.replace('+0000 UTC', ''))
uber_data[4] = uber_data[4].map(lambda x: x.replace('+0000 UTC', ''))
uber_data[8] = uber_data[8].map(lambda x: x.replace('+0000 UTC', ''))
uber_data
```

```python
uber_data[3] = pd.to_datetime(uber_data[3])
uber_data[4] = pd.to_datetime(uber_data[4])
uber_data[8] = pd.to_datetime(uber_data[8])
uber_data
```

```python
df_final = uber_data
df_final.columns = ['city', 'trip status', 'request time', 'start time', 'end time', 'distance(mi)', 'fare', 'currency']
df_final
```

```python
# converting datatype from object to numeric for distance
df_final['distance(mi)'] = pd.to_numeric(df_final['distance(mi)'])
df_final.dtypes
```

```python
# at this point we can also drop trip status and currency columns
# as these won't be used for analysis and we have also filtered out
# any trips that were not complete as well as set the currency for all trips
# to GBP

df_final.drop(['trip status', 'currency'], axis = 1, inplace = True)
df_final
```

```python
# calculating average distance travelled per year
avg_distance = df_final.groupby([df_final['start time'].dt.year])['distance(mi)'].mean()
avg_distance
```

```python
# plotting a bar chart of average distance travelled each year
avg_distance.plot(kind='bar', figsize=(8,6))
plt.ylabel('avg_distance')
plt.xlabel('year')
plt.title('avg distance(mi) / year')
```

```python
# calculating average amount of money spent on trips per year
# method used for this was the same as the average distance travelled calculation
avg_spent = df_final.groupby([df_final['start time'].dt.year])['fare'].mean()
avg_spent
```

```python
# plotting a bar chart of average money spent each year
avg_distance.plot(kind='bar', figsize=(8,6))
plt.ylabel('avg_spent(GBP)')
plt.xlabel('year')
plt.title('avg spent / year')
```

```python
#calculate number of trips per month
trips_per_month = df_final.groupby([df_final['start time'].dt.month])['start time'].count()
trips_per_month
```

```python
# plot another bar chart showing number of trips per month
trips_per_month.plot(kind='bar', figsize=(8,6))
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.ylabel('number of trips')
plt.xlabel('month')
plt.title('trips / month')
```
